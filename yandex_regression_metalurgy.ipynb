# Предсказание температуры стали

Чтобы оптимизировать производственные расходы, металлургический комбинат решил уменьшить потребление электроэнергии на этапе обработки стали.

Требуется построить модель, прогнозирующую температуру стали. Это позволит снизить затраты на электричество и обслуживание.  
Модель должна предсказывать температуру с погрешностью менее 6.8 градусов (**MAE < 6.8**).

## Описание этапа обработки

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша.

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (от tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (от slab, «плита»).

Подробнее о процессе внепечной обработки металлов можно почитать [здесь](https://ppt-online.org/497835).

## Описание данных

Данные состоят из файлов, полученных из разных источников:
- `data_arc_new.csv` — данные об электродах;

- `data_bulk_time_new.csv` — данные о подаче сыпучих материалов (время);

Во всех файлах столбец `key` содержит номер партии. В файлах может быть несколько строк с одинаковым значением `key`: они соответствуют разным итерациям обработки.

## Исследование признаков

```python
def IQR_borders(series, coef=1.5, skip_zero=False):
    if skip_zero:
        series = series.replace({0.0: np.nan})
    
    IQR = series.quantile(0.75) - series.quantile(0.25)
    low_border = series.quantile(0.25) - coef * IQR
    top_border = series.quantile(0.75) + coef * IQR
    
    if low_border < series.min():
        low_border = series.min()
    if top_border > series.max():
        top_border = series.max()
    
    return low_border, top_border
```

1. Существенный вклад внесло создание такого признака, как суммарная работа по нагреву. Первое измерение температуры также очень важно. Возможно, искуственное создание большего числа признаков также повысит точность.
2. Масштабирование данных также должно было оказать значительное влияние. В противном случае модель могла быть "решить", что такой признак, как общее время обработки, более важен, чем прочие. Хотя анализ важности признаков показал обратное.

- Признаки, которые были использованы для обучения и их предобработка
1. **Суммарная работа** для каждой партии была рассчитана как 
$\sum_{i=1}^{n} \text{Время нагрева }_i \cdot \sqrt{\text{Активная мощность}_i^2 + \text{Реактивная мощность}_i^2}$  
, где i – номер нагрева дугой, а n – число нагревов дугой для партии.
2. **Общее время обработки** для каждой партии было рассчитано как 
$\text{Конец нагрева дугой }_n -  \text{Начало нагрева дугой }_1$  
, где n – число нагревов дугой для партии.
3. Были использованы объёмы добавленных сыпучих материалов 3, 4, 6, 12, 14 и 15, а также объёмы добавленных проволочных материалов 1 и 2. Пропуски в этих данных были заполнены нулями.
4. Из данных о температуре были удалены партии с пропусками в значениях температуры. Также были удалены партии с одним измерением температуры. Среди оставшихся данных для каждой партии первое измерение температуры было принято за начальное (то есть за признак для обучения), последнее – за конечное (то есть за целевую метрику).
